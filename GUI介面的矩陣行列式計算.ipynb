```python
def calculation():
    try:
        e11 = eval(entry11.get())
        e12 = eval(entry12.get())
        e13 = eval(entry13.get())
        e21 = eval(entry21.get())
        e22 = eval(entry22.get())
        e23 = eval(entry23.get())
        e31 = eval(entry31.get())
        e32 = eval(entry32.get())
        e33 = eval(entry33.get())
        ans = e11*e22*e33 + e12*e23*e31 + e13*e21*e32 - e13*e22*e31 - e12*e21*e33 - e11*e23*e32  
        label.config(text=ans)
    except:
        messagebox.showerror("Error","請輸入完整矩陣！")
    
def clear():
    entry11.delete(0, "end")
    entry12.delete(0, "end")
    entry13.delete(0, "end")
    entry21.delete(0, "end")
    entry22.delete(0, "end")
    entry23.delete(0, "end")
    entry31.delete(0, "end")
    entry32.delete(0, "end")
    entry33.delete(0, "end")
    label.config(text="計算結果")

    
import tkinter as tk
from tkinter import messagebox

window = tk.Tk()
window.iconbitmap('logo.ico')
window.title('Matrix determinant computer')
window.geometry('400x200')

messagebox.showinfo("Information","歡迎使用矩陣行列式計算機！")

frame0 = tk.Frame(window)
frame0.pack(padx=5, pady=5)
frame1 = tk.Frame(window)
frame1.pack(padx=5, pady=5)
frame2 = tk.Frame(window)
frame2.pack(padx=5, pady=5)

menubar = tk.Menu(window)

filemenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='File', menu=filemenu)
filemenu.add_command(label='Exit', command=window.destroy)

editmenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='Edit', menu=editmenu)
editmenu.add_command(label='clear', command=clear)        

window.config(menu=menubar)

entry11 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry11.config(width=10)
entry11.grid(column=0, row=0)
entry12 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry12.config(width=10)
entry12.grid(column=1, row=0)
entry13 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry13.config(width=10)
entry13.grid(column=2, row=0)

entry21 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry21.config(width=10)
entry21.grid(column=0, row=1)
entry22 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry22.config(width=10)
entry22.grid(column=1, row=1)
entry23 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry23.config(width=10)
entry23.grid(column=2, row=1)

entry31 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry31.config(width=10)
entry31.grid(column=0, row=2)
entry32 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry32.config(width=10)
entry32.grid(column=1, row=2)
entry33 = tk.Entry(frame0, bg="#4B8BBE", font="JhengHei 15", borderwidth=3)
entry33.config(width=10)
entry33.grid(column=2, row=2)

label = tk.Label(frame1, bg="#FFE873", font="Helvetica 15 bold", anchor="e", text="計算結果")
label.config(width=25)
label.grid(column=0, row=0)

calculation = tk.Button(frame2, text="計算", command=calculation)
calculation.config(width=15)
calculation.grid(column=0, row=0)

clear = tk.Button(frame2, text="清除", command=clear)
clear.config(width=15)
clear.grid(column=1, row=0)

window.mainloop()
```
